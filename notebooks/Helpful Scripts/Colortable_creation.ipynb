```python
pal = 'Vortex BV.pal'
```

```python
"""
This script allows you to use .pal files to create new colormaps in python. All that is needed is the input file

"""



def read_pal(pal):
    results = []
    with open(pal, 'r') as file:
        for line in file:
            line = line.strip()
            if not line:
                continue

            tokens = line.split()

            # Try parsing numeric rows
            try:
                if len(tokens) >= 4:
                    value = float(tokens[0])
                    r, g, b = map(int, tokens[1:4])
                elif len(tokens) == 3:
                    value = len(results)  # auto index if no values
                    r, g, b = map(int, tokens)
                else:
                    continue
            except:
                continue

            hex_color = "#{:02x}{:02x}{:02x}".format(r, g, b)
            results.append((value, hex_color))

    return dict(results)
            
def colortable_register(pal_file):
    cmap_name = input("Choose a name for your colormap: ")
    hex_dict = read_pal(pal_file)

    if not hex_dict:
        raise ValueError("No valid color rows found in .pal file.")

    min_val, max_val = min(hex_dict), max(hex_dict)

    normalized = [
        (val - min_val) / (max_val - min_val)
        for val in hex_dict
    ]

    sorted_items = sorted(zip(normalized, hex_dict.values()))
    colors = [color for _, color in sorted_items]

    cmap = mcolors.LinearSegmentedColormap.from_list(cmap_name, colors)
    plt.colormaps.register(cmap=cmap)

    print(f"Registered colormap: {cmap_name}")
```

```python
colortable_register(pal)
```
